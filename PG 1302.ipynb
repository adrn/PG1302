### Normalize the data by subtracting the median

```python
t = css['MJD']
y = css['Mag'] - np.mean(css['Mag'])
dy = css['Magerr']
```

### Run the astroml implementations of the Lomb-Scargle periodogram, with and without uncertainties

```python
omega = np.logspace(-3,-1,1000)
f = omega / 2 / np.pi

P_sta = lomb_scargle(t, y, dy, omega, generalized=False, subtract_mean=False) # standard
P_gen = lomb_scargle(t, y, dy, omega, generalized=True, subtract_mean=False) # generalized

peak_omega = omega[P_gen.argmax()]
peak_freq = f[P_gen.argmax()]

peak2_omega = omega[f > 1E-3][P_gen[f > 1E-3].argmax()]
peak2_freq = peak2_omega / 2 / np.pi
```

```python
plt.figure(figsize=(12,8))
plt.semilogx(f, P_sta, marker=None, label="Standard")
plt.semilogx(f, P_gen, marker=None, label="Generalized")
plt.axvline(peak_freq)
plt.xlim(f.min(), f.max())
plt.legend(fontsize=22)
plt.xlabel("Frequency [day$^{-1}$]")
```

### Estimate error due to sampling

```python
niter = 100

y_shuff = y.copy()
dy_shuff = dy.copy()

P_shuf = np.zeros_like(omega)
ix = np.arange(len(y)).astype(int)
for i in range(niter):
    np.random.shuffle(ix)
    P_shuf += lomb_scargle(t, y[ix], dy[ix], omega, generalized=True)
```

```python
plt.figure(figsize=(12,8))
#plt.semilogx(f, P_gen, marker=None)
plt.semilogx(f, P_shuf / float(niter), marker=None)
plt.xlim(f.min(), f.max())
plt.xlabel("Frequency [day$^{-1}$]")
```

### Compute window function and estimate PSD of window function

The window function will be a bunch of top-hats centered on the observation times specified.

```python
nbins = 2**15

t_window = np.linspace(t.min(), t.max(), nbins)

D = abs(t[:, np.newaxis] - t_window)
i = np.argmin(D, axis=1)

window = np.zeros(nbins)
window[i] = 1

plt.figure(figsize=(10,6))
plt.title("Window Function")
plt.plot(t_window, window, drawstyle='steps', marker=None)
plt.xlim(t_window.min(), t_window.max())
plt.ylim(-0.02,1.1)
```

```python
P_window = lomb_scargle(t_window, window, 1, omega,
                        generalized=False, subtract_mean=False)
```

```python
plt.figure(figsize=(12,8))
plt.semilogx(f, P_gen, marker=None)
plt.semilogx(f, P_window*100, marker=None)
plt.xlim(f.min(), f.max())
plt.xlabel("Frequency [day$^{-1}$]")
```

### Estimate confidence bounds

```python
D = lomb_scargle_bootstrap(t, y, dy, omega, generalized=True,
                           N_bootstraps=1000, random_state=0)
sig1, sig5 = np.percentile(D, [99, 95])
```

```python
plt.figure(figsize=(12,8))
plt.semilogx(f, P_gen, marker=None)
plt.axhline(sig1, linestyle='dashed')
plt.axhline(sig5, linestyle='dashed')
plt.xlim(f.min(), f.max())
plt.xlabel("Frequency [day$^{-1}$]")
```

### Interesting, some very significant peaks there...

Now I'll fold the light curve on the first period

```python
def time_to_phase(time, period, t0=0.):
    """ Convert an array of times in days to an array of phases """
    return ((time-t0) % period) / period
```

```python
phase = time_to_phase(t, 1./peak_freq)
```

Fit for coefficient and phase of sinusoid...

```python
def model(p, omega, t):
    a,phi = p
    return a*np.cos(omega*t + phi)

def func(p, t, y, dy, omega):
    return (y - model(p,omega,t))**2 / dy**2

x0 = [0.2, 0.]
xbest,ier = so.leastsq(func, x0, args=(t,y,dy,peak_omega))
ier
```

```python
ttt = np.linspace(t.min(),t.max(),10000)

plt.figure(figsize=(12,8))
plt.errorbar(phase, y, dy, marker='o', linestyle='none', ecolor='#777777', alpha=0.5)

phase_ttt = time_to_phase(ttt, period=1/peak_freq)
plt.plot(phase_ttt, model(xbest, peak_omega, ttt), linestyle='none')

plt.title("Folded light curve")
plt.xlabel("Phase")
plt.ylabel(r"$\Delta {\rm mag}$")
```

```python
ttt = np.linspace(t.min(),t.max(),10000)

plt.figure(figsize=(12,8))
plt.errorbar(phase, y - model(xbest, peak_omega, t), dy, 
             marker='o', linestyle='none', ecolor='#777777', alpha=0.5)

plt.ylim(-0.2,0.2)
plt.title("Folded, subtracted light curve")
plt.xlabel("Phase")
plt.ylabel(r"$\Delta {\rm mag}$")
```
